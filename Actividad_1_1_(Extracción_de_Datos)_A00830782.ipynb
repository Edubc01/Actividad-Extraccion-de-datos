##Filtro 8
Filtro de conectores logicos

Cantidades menores a 2 en almacen y durante las fechas 05/01/2022 00:00

```python
filtro8 = df[(df["CANT"] < 2) & (df["FECHA_DOC"] == "05/01/2022 00:00")]
filtro8
```

```python
filtro8.to_csv("FILTRO8.csv")
from google.colab import files
files.download("FILTRO8.csv")
```

##Filtro 9
Filtro not

Cantidades mayores a 1 en almacen y que no sea en las fechas 05/01/2022 00:00

```python
filtro9 = df[~(df["CANT"] >= 2) & ~(df["FECHA_DOC"] == "05/01/2022 00:00")]
filtro9
```

```python
filtro9.to_csv("FILTRO9.csv")
from google.colab import files
files.download("FILTRO9.csv")
```
